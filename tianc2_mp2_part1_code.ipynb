```python
import numpy as np
import os
import time
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from skimage import io, color, transform
import scipy
from scipy.ndimage.filters import convolve
from scipy.ndimage.filters import gaussian_laplace, generic_filter, gaussian_filter
from numpy.random import rand, randint
from matplotlib.patches import Circle
```

```python
IMG_DIR = 'C:/Users/n551/Desktop/ECE549/MP/MP2/'
IMG_NAME = 'c1.jpg'
```

```python
def read_image(IMG_NAME):
    # YOUR CODE HERE
    img = mpimg.imread(IMG_DIR+IMG_NAME)
    return img
```

```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
img = read_image(IMG_NAME)
gray1 = rgb2gray(img)
plt.figure(figsize=(10,6))
plt.imshow(gray1,cmap='Greys_r')
plt.show()
```

```python
IMG_NAME = 'c2.jpg'
img = read_image(IMG_NAME)
gray2 = rgb2gray(img)
plt.figure(figsize=(10,6))
plt.imshow(gray2,cmap='Greys_r')
plt.show()
```

```python
imput = scipy.signal.unit_impulse((5, 5), 'mid')
gaussian_kernel = scipy.ndimage.gaussian_filter(imput,sigma=1)
gaussian_kernel = 1.5*imput - gaussian_kernel
# print(gaussian_kernel)
change1 = scipy.signal.convolve2d(gray1,gaussian_kernel,mode='same')
change1 = np.float32(change1)
plt.figure(figsize=(10,6))
plt.imshow(change1,cmap='Greys_r')
plt.show()
```

```python
change2 = scipy.ndimage.gaussian_filter(gray2,sigma=2)
change2 = np.float32(change2)
plt.figure(figsize=(10,6))
plt.imshow(change2,cmap='Greys_r')
plt.show()
```

```python
result2 = change1 + change2
plt.figure(figsize=(10,6))
plt.imshow(result2,cmap='Greys_r')
plt.show()
```
